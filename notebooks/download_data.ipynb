```python
import os
import sys
import kagglehub
import matplotlib.pyplot as plt
sys.path.append('../')
from src.utils import get_path_to

os.environ['KAGGLEHUB_CACHE'] = get_path_to('input')
```

```python
path = kagglehub.dataset_download("igorkarayman/signatures-and-stamps")

print("Path to dataset files:", path)
```

```python
def visualize_images(
    path: str,
    n_images: int
) -> None:
    image_found = False
    while not image_found:
        inside_folder = os.listdir(path)
        if '.jpg' in inside_folder[0]:
            image_found = True
            break
        path = os.path.join(path, inside_folder[0])

    images = [f for f in os.listdir(path) if f.endswith('.jpg')]
    images = images[:n_images]
    images = [os.path.join(path, f) for f in images]

    images = [plt.imread(image) for image in images]

    # Visualize the images
    n_cols = min(3, len(images))
    n_rows = (len(images) + n_cols - 1) // n_cols  # Ceiling division
    plt.figure(figsize=(10, 10))
    for i, image in enumerate(images):
        plt.subplot(n_rows, n_cols, i+1)
        plt.imshow(image)
        plt.axis('off')
    plt.show()
```

```python
visualize_images(path=path, n_images=12)
```
